### [문제 2] 범죄 데이터프레임 정리

```python
# 필요한 모듈 임포트
import numpy as np   # 수치 계산을 위해 numpy 임포트, numpy를 np라고 부르겠다는 의미
import pandas as pd  # 데이터 분석 및 조작을 위해 pandas 임포트, pandas를 pd라고 부르겠다는 의미
```

```python
# crime_2019.txt 파일을 읽어서 crime_2019 데이터프레임에 저장
crime_2019 = pd.read_table('crime_2019.txt',
                            sep = '\t',        # 데이터 구분자는 탭('\t')으로 지정
                            encoding='utf-8')  # 인코딩 utf-8

crime_2019.head()
```

```python
# 열 이름 수정, 원본 수정
crime_2019.rename(columns={crime_2019.columns[4] : '살인',
                           crime_2019.columns[5] : '살인 검거',
                           crime_2019.columns[6] : '강도',
                           crime_2019.columns[7] : '강도 검거',
                           crime_2019.columns[8] : '강간',
                           crime_2019.columns[9] : '강간 검거',
                           crime_2019.columns[10] : '절도',
                           crime_2019.columns[11] : '절도 검거',
                           crime_2019.columns[12] : '폭력',
                           crime_2019.columns[13] : '폭력 검거'}, inplace=True)

# 자치구 열을 인덱스로 지정, 원본 수정
crime_2019.set_index('자치구', inplace=True)

# 불필요한 데이터 행과 열 삭제
crime_2019.drop(['자치구', '합계'], inplace=True)  # 자치구행과 합계행 삭제, 원본 수정
del crime_2019['기간']                             # 기간 열 삭제
del crime_2019['합계']                             # 합계 열 삭제
del crime_2019['합계.1']                           # 합계.1 열 삭제

crime_2019.head()
```

```python
# '강간 검거', '절도', '절도 검거', '폭력', '폭력 검거' 열의 데이터에는 천단위 구분 쉼표가 포함되어 있음
# 작업을 위해 천단위 구분 쉼표 제거
crime_2019['강간 검거'] = crime_2019['강간 검거'].str.replace(',', '')
crime_2019['절도'] = crime_2019['절도'].str.replace(',', '')
crime_2019['절도 검거'] = crime_2019['절도 검거'].str.replace(',', '')
crime_2019['폭력'] = crime_2019['폭력'].str.replace(',', '')
crime_2019['폭력 검거'] = crime_2019['폭력 검거'].str.replace(',', '')

crime_2019.head()
```

```python
# 검거율 계산 작업을 위해 데이터 타입 확인
crime_2019.dtypes
```

```python
# 데이터 타입을 숫자로 변경하고 에러처리 방식 지정, 결측치는 0으로 대체
crime_2019 = crime_2019.apply(pd.to_numeric, errors='coerce').fillna(0)
crime_2019.dtypes
```

```python
# 검거율 계산 -> 검거율 = 검거 / 발생 * 100
crime_2019['살인검거율'] = crime_2019['살인 검거'] / crime_2019['살인'] * 100
crime_2019['강도검거율'] = crime_2019['강도 검거'] / crime_2019['강도'] * 100
crime_2019['강간검거율'] = crime_2019['강간 검거'] / crime_2019['강간'] * 100
crime_2019['절도검거율'] = crime_2019['절도 검거'] / crime_2019['절도'] * 100
crime_2019['폭력검거율'] = crime_2019['폭력 검거'] / crime_2019['폭력'] * 100

# 검거 열 모두 제거
del crime_2019['살인 검거']
del crime_2019['강도 검거']
del crime_2019['강간 검거']
del crime_2019['절도 검거']
del crime_2019['폭력 검거']

crime_2019.head()
```

```python
# 이전에 작업했던 pop_cctv_2018.csv' 파일에서 자치구와 2018년도 인구데이터, CCTV 데이터만 읽어와서 pop_cctv_2018 데이터 프레임에 적재
pop_cctv_2018 = pd.read_csv('./pop_cctv_2018.csv', usecols = ['자치구', '인구 규모 2018', 'CCTV 규모 2018'])

# 열 이름 변경, 원본 수정
pop_cctv_2018.rename(columns={pop_cctv_2018.columns[1] : '인구 규모',  # '인구 규모 2018'열을 '인구 규모'로
                              pop_cctv_2018.columns[2] : 'CCTV 규모'}, inplace=True)  # 'CCTV 규모 2018'열을 'CCTV 규모'로

pop_cctv_2018
```

```python
# '자치구' 열을 key 열로 지정하여 crime_2019 데이터프레임과 pop_cctv_2018 데이터프레임을 병합하고 crime_2019로 저장
# 옵션을 outer로 지정하여 두 데이터프레임에 존재하는 모든 데이터를 합친다. 
crime_2019 = pd.merge(crime_2019, pop_cctv_2018, on='자치구', how='outer')

crime_2019.set_index('자치구', inplace=True)  # 자치구 열을 인덱스로 지정, 원본 수정

crime_2019 # <그림 D2>
```

### [문제 3] 범죄 발생 건수 정규화

```python
# 위 표를 보면 절도와 폭력은 천 단위인데, 강간은 백 단위이고, 강도와 살인은 십 단위로 범죄 5종의 수치가 각기 다르다는 것을 알 수 있음
# 이 수치들을 모두 (0,1) 범위로 정규화

from sklearn import preprocessing  # scikit-learn 임포트

# col 열에 대하여 crime_2019로부터 crime_2019_norm으로 정규화 변환하여 복사 
col = ['살인', '강도', '강간', '절도', '폭력']
crime_2019_norm = pd.DataFrame(
                  index=crime_2019.index,
                  columns=col, 
                  data=preprocessing.MinMaxScaler().fit_transform(crime_2019[col].values.astype(float))).round(2)
                  # 데이터 타입은 float로 지정하고, 소수점 아래 둘째자리까지 출력

crime_2019_norm  # <그림 D3>
```

### [문제 4] 범죄 발생 비율 계산

```python
# crime_2019_per_capita 데이터 프레임 생성
# 인덱스는 crime_2019 데이터프레임의 인덱스로 지정하고, '살인', '강도', '강간', '절도', '폭력' 열 생성
crime_2019_per_capita = pd.DataFrame(index=crime_2019.index, columns=['살인', '강도', '강간', '절도', '폭력'])

# crime_2019 데이터프레임의 범죄 건수와 인구 규모를 활용 하여 십만 명 당 범죄 발생 비율을 계산하여 각 열에 저장
crime_2019_per_capita['살인'] = crime_2019['살인'] / crime_2019['인구 규모'] * 100000
crime_2019_per_capita['강도'] = crime_2019['강도'] / crime_2019['인구 규모'] * 100000
crime_2019_per_capita['강간'] = crime_2019['강간'] / crime_2019['인구 규모'] * 100000
crime_2019_per_capita['절도'] = crime_2019['절도'] / crime_2019['인구 규모'] * 100000
crime_2019_per_capita['폭력'] = crime_2019['폭력'] / crime_2019['인구 규모'] * 100000

# 데이터는 소수점 둘째자리까지 반올림
crime_2019_per_capita = crime_2019_per_capita.round(2)
crime_2019_per_capita  # <그림 D4>
```

```python
station_addreess = []  # 획득한 경찰서의 주소를 저장
station_lat = []       # 획득한 경찰서의 위도를 저장
station_lng = []       # 획득한 경찰서의 경도를 저장
    
for name in station_name:                                    # station_name(관서명)의 내부 항목을 name으로 반복
    tmp = gmaps.geocode(name, language='ko')                 # 구글 지도에서 name 검색 -> tmp라는 임시변수에 저장
    station_addreess.append(tmp[0].get("formatted_address")) # 구글 지도에서 주소를 획득
    
    # tmp[0]의 이미는 정상적이라면 주소는 하나만 넘어와야 하지만 
    # 혹시나 주소가 여러개일 경우, 그 중에 제일 앞에 있는 값 하나만 가져오라는 의미
    
    tmp_loc = tmp[0].get("geometry")                         # 구글 지도에서 "geometry" 획득 -> tmp_loc라는 변수에 저장
 
    station_lat.append(tmp_loc['location']['lat'])           # 구글 지도에서 위도 획득 (geometry의 location 안의 lat 값)
    station_lng.append(tmp_loc['location']['lng'])           # 구글 지도에서 경도 획득 (geometry의 location 안의 lng 값)
    
    print(f'{name:<12} \t {tmp[0].get("formatted_address")}')  # 획득한 주소 출력
    
office['위도'] = station_lat  # office 데이터프레임의 '위도'열에 위도 데이터 저장
office['경도'] = station_lng  # office 데이터프레임의 '경도'열에 경도 데이터 저장
```

```python
# 획득한 주소에서 자치구 이름을 추출해서 `자치구` 열에 저장

# station_address 리스트에서 구 이름만 추출하여 gu_name 리스트에 저장
# -> 각 경찰서가 어느 자치구에 위치해있는지 파악하기 위함

gu_name = []

for address in station_addreess:                             # station_address(주소)의 내부 항목을 name으로 반복
    tmp = address.split()                                     # 문자열을 분리하는 split() 함수 -> 주소를 띄어쓰기 단위로 쪼갠 후
    tmp_gu = [word for word in tmp if word[-1] == '구'][0]  #  끝이 '구'로 끝나는 데이터만 가져와서 자치구 데이터 추출
    print(f'tmp_gu: {tmp_gu:<5} \t tmp: {tmp}')               # 획득한 값 출력
    gu_name.append(tmp_gu)                                    # tmp_gu에 저장되어있는 자치구 이름 데이터를 gu_name에 저장
    
office['자치구'] = gu_name  # office 데이터프레임의 '자치구'열에 자치구 이름 데이터 저장
```
